```python
import datetime

saldo = 1000
nsaldodiario = 0
operacoes = []
ultima_atualizacao = None

menu='''Bem-vindo ao sistema bancário!
Selecione uma das opções, digitando seu respectivo número:
1 - Depósito
2 - Saque
3 - Extrato
4 - Sair
=>'''

while True:
    opcao=int(input(menu))
    if opcao==1:
        print("===== Depósito =====")
        deposito = input("Por favor, digite o valor do deposito")
        while True:
            try:
                deposito = float(deposito)
                if deposito > 0:
                    break
                else:
                    print("Por favor, digite um valor positivo maior que zero")
            except ValueError:
                print("Por favor, digite um valor numérico válido")
            deposito = input("Por favor, digite o valor do deposito:") 
        deposito = float(deposito)
        saldo += deposito
        operacoes.append(f"Depósito feito de {deposito:.2f}")
    elif opcao == 2:
        print("===== Saque =====")
        data_atual = datetime.date.today()
        if ultima_atualizacao is None or data_atual > ultima_atualizacao:
            nsaldodiario = 0 
            ultima_atualizacao = data_atual
        
        if nsaldodiario < 3:
            saque = input("Digite o valor do saque:")
            
            while True:
                try:
                    saque = float(saque)
                    if saque > 0:
                        break
                    else:
                        print("Por favor, digite um valor positivo maior que zero")
                except ValueError:
                    print("Por favor, digite um valor numérico válido")
                saque = input("Por favor, digite o valor do saque:")
                                 
            saque = float(saque)
            if saque <= saldo:
                saldo -= saque
                operacoes.append(f"Saque feito de {saque:.2f}")
                nsaldodiario += 1
            else:
                print("Saldo insuficiente!")
        else:
            print("Você atingiu o limite de saque diário! Não é possível mais realizar essa operação hoje!")    
    elif opcao == 3:
        print("===== Extrato =====")
        if operacoes:
            for operacao in operacoes:
                print(operacao)
        else:
            print("Não há operações realizadas.")
        print(f"Saldo atual: R$ {saldo:.2f}")
        print("=================")
    elif opcao == 4:
        print("Obrigado por usar o sistema bancário!")
        break
    else:
        print("Opção inválida. Por favor, selecione uma opção válida.")
                
                
                
            
```
