```python
import textwrap

def menu():
    menu='''Bem-vindo ao sistema bancário!
Selecione uma das opções, digitando seu respectivo número:
1 - Depósito
2 - Saque
3 - Extrato
4 - Nova conta
5 - Listar contas
6 - Novo usuário
7 - Sair
=>'''
    return input(textwrap.dedent(menu)).strip()

def deposito(saldo, valor, extrato):
    if valor > 0:
        saldo += valor
        extrato += f"Depósito realizado de {valor:.2f}\n"  # Adiciona a operação ao extrato
        print(f"Depósito feito de {valor:.2f}")
    else:
        print("Por favor, digite um valor positivo maior que zero")
        
    return saldo, extrato
    
def saque(*, saldo, valor, extrato, limite, número_saques, limite_saques):
    if número_saques >= limite_saques: 
        print("Limite diário de saques excedido! Tente novamente outro dia!")
    else:
        if valor > limite:
            print("Limite ultrapassado!")
        elif valor > saldo:
            print("Saldo insuficiente!")
        else:
            saldo -= valor
            extrato += f"Saque realizado de {valor:.2f}\n"  # Adiciona a operação ao extrato
            print(f"Saque realizado de {valor:.2f}")
            número_saques += 1
    return saldo, extrato, número_saques

def exibir_extrato(saldo, /, *, extrato):
    print("Não há operações realizadas." if not extrato else extrato)
    print(f"Saldo atual: R$ {saldo:.2f}")

# Lista de usuários
usuarios = []

# Lista de contas
contas = []

# Próximo número de conta
próximo_número_conta = 1

def filtrar_usuario(cpf, usuario):
    usuarios_filtrados=[usuario for usuario in usuarios if usuario['cpf'] == cpf]
    return usuarios_filtrados[0] if usuarios_filtrados else None

def cadastrar_usuario(nome, data_nascimento, cpf, endereco):
    """Cadastra um novo usuário."""
    usuario = filtrar_usuario(cpf, usuarios)
    if usuario:
        print("\n Já existe usuário com esse CPF!\n")
        return
    
    usuarios.append({
        'nome': nome,
        'data_nascimento': data_nascimento,
        'cpf': cpf,
        'endereco': endereco
    })
    print("Usuário cadastrado com sucesso!")

def cadastrar_conta(agencia, numero_conta, usuario):
    """Cadastra uma nova conta."""
    contas.append({
        'agencia': agencia,
        'numero_conta': numero_conta,
        'usuario': usuario
    })
    print("Conta cadastrada com sucesso!")

def criar_conta():
    """Cria uma nova conta para um usuário existente."""
    cpf = input("Digite o CPF do usuário: ")
    usuario = None
    for user in usuarios:
        if user['cpf'] == cpf:
            usuario = user
            break
    if usuario:
        agencia = '0001'
        global próximo_número_conta
        numero_conta = próximo_número_conta
        próximo_número_conta += 1
        cadastrar_conta(agencia, numero_conta, usuario)
    else:
        print("Usuário não encontrado.")

def listar_contas():
    """Lista todas as contas cadastradas."""
    if contas:
        for conta in contas:
            print(f"Agência: {conta['agencia']}, Conta: {conta['numero_conta']}, Titular: {conta['usuario']['nome']}")
    else:
        print("Não há contas cadastradas.")

def loop_principal():
    """Função para o loop principal do sistema."""
    saldo = 0
    extrato = ""
    limite = 500
    número_saques = 0
    limite_saques = 3

    while True:
        opcao = menu()

        if opcao == '1':
            valor = float(input("Digite o valor do depósito: "))
            saldo, extrato = deposito(saldo, valor, extrato)

        elif opcao == '2':
            valor = float(input("Digite o valor do saque: "))
            saldo, extrato, número_saques = saque(saldo=saldo, valor=valor, extrato=extrato, limite=limite, número_saques=número_saques, limite_saques=limite_saques)

        elif opcao == '3':
            exibir_extrato(saldo, extrato=extrato)

        elif opcao == '4':
            criar_conta()

        elif opcao == '5':
            listar_contas()

        elif opcao == '6':
            nome = input("Digite o nome do usuário: ")
            data_nascimento = input("Digite a data de nascimento (dd/mm/aaaa): ")
            cpf = input("Digite o CPF: ")
            endereco = input("Digite o endereço (logradouro, nro - bairro - cidade/sigla estado): ")
            cadastrar_usuario(nome, data_nascimento, cpf, endereco)

        elif opcao == '7':
            print("Saindo do sistema...")
            break

        else:
            print("Opção inválida! Tente novamente.")

# Chama a função loop_principal
loop_principal()
```
